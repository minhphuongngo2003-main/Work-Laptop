```python
from vnstock import Vnstock
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from vnstock import Company
from vnstock import Finance
import plotly. express as px
```

```python
ticker = 'MBB'

stock = Vnstock().stock(symbol= ticker, source='VCI') 

company = Company( symbol = ticker, source = 'VCI')

finance = Finance (symbol = ticker, source = 'VCI')

df = stock.quote.history(start = '2024-08-29', end = '2025-08-29', interval = '1D')

df = df.drop(columns = ['high', 'low', 'volume', 'open'])

df['time'] = pd.to_datetime (df ['time'])

df.sort_values(by ='time',ascending=False,inplace=True)

df
```

```python
df_statement = finance.income_statement (period = 'quarter', lang = 'en')

df_statement
```

```python
df_statement_only = df_statement[['ticker', 'yearReport', 'lengthReport', 'Attributable to parent company']]

df_statement_only
```

```python
df_ratio = finance.ratio (period = 'quarter', lang = 'vi')

df_ratio
```

```python
df_ratio_only = df_ratio [[('Meta', 'CP'), ('Meta', 'Năm'), ('Meta', 'Kỳ'), ('Chỉ tiêu định giá', 'Số CP lưu hành (Triệu CP)')]]
df_ratio_only.columns = df_ratio_only.columns.map('_'.join)  
df_statement_only
```

```python
df_ratio_only.columns = ['ticker', 'yearReport','lengthReport','outstanding_share']
```

```python
df_statement_only.columns
```

```python
df_ratio_only['Period_'] = df_ratio_only.yearReport.astype(str) + '_' + df_ratio_only.lengthReport.astype(str)
df_statement_only['Period_'] = df_statement_only.yearReport.astype(str) + '_' + df_statement_only.lengthReport.astype(str)
```

```python
DF_merge = pd.merge(df_ratio_only,df_statement_only,on='Period_',how='inner')
DF_merge = DF_merge[['outstanding_share','Attributable to parent company','Period_']]
```

```python
DF_merge.rename(columns={'Attributable to parent company':'Profit_Shareholder'},inplace=True)
DF_merge.sort_index(ascending=False,inplace=True)
```

```python
DF_merge.outstanding_share = DF_merge.outstanding_share.iloc[-1]
DF_merge['TMT_Profit'] = DF_merge.Profit_Shareholder.rolling(window=4, min_periods=4).sum()
DF_merge
```

```python
DF_merge.set_index('Period_')
```

```python
DF_merge['EPS'] = DF_merge.TMT_Profit/DF_merge.outstanding_share
# DF_merge['Applied_Date'] = 
```

```python

years = df_ratio_only[('Meta', 'Năm')]

pe = df_ratio_only [('Chỉ tiêu định giá', 'P/E')]

plt.plot (years, pe, marker = 'o')
plt.xlabel("Năm")
plt.ylabel('P/E')
plt.title("P/E qua các năm")
plt.grid(True)
plt.show()
```

```python
fig = px.line(
    x = years,
    y = pe, 
    markers = True,
    labels = {"x": 'Năm', 'y' : 'P/E'},
    title = "P/E qua các năm"
)

fig.show()


```
